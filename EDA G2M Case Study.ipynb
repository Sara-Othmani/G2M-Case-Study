```python
#cab data Univariate Analysis
import seaborn as sns
sns.countplot(x='Company',data=cab) 
## Looking at the results, we can tell that the majority of the cab companies are the Yellow Cab

cab['Company'].value_counts()
## The exact breakdown of the companies are : 
## 274681 Yellow cabs VERSUS 84711 Pink cabs
## Which means that the number of yellow cabs is approximately about 3X pink cabs
```

```python
# Box plot to visualize the relationship between KM Travelled and Price Charged
sns.boxplot(x="KM Travelled", y="Price Charged", data = cab)
#Box plot to visualize the relationship between KM Travelled and Cost of Trip
sns.boxplot(x="KM Travelled", y="Cost of Trip", data = cab)
## As many more KM are travelled, the price charged & the cost of the trip are both proportionally higher.

# Assumption: The three variables : 'KM Travelled' 'Price Charged' 'Cost of Trip' are highly correlated. 
#Verify assumption

# Correlation plot // Relation between variables
plt.figure(figsize=(10,20))
c = cab.corr()
sns.heatmap(c, cmap="BrBG", annot=True)
c
## 'KM Travelled' & 'Price Charged', 'KM Travelled' and 'Cost of Trip' are too couples of variables that are highly correlated.

#Correlation between the charge/cost and date of travel
sns.boxplot(data=cab, x='Cost of Trip', y='Date of Travel')
```

```python
# Customer dataset
# Transaction dataset

# Statistical data analysis
# Summary statistics
customer.describe()
transaction.describe()

# Combine the two datasets 'transaction' and 'customer', as both of them have 'Customer ID' column 
# Use merge() function
customer_transaction = pd.merge(customer, transaction)
customer_transaction
```

```python
#customer_transaction new merged dataset analysis

#univariate analysis

import seaborn as sns
sns.countplot(x='Gender', data = customer_transaction) 

## There is no equal distribution Gender wise : Male customers > Female customers

## The exact breakdown of the companies are : 
customer_transaction['Gender'].value_counts()
## 256611 Male cab users VERSUS 183487 Female cab users
```

```python
#Make an all-in-one dataset 

# Create a new dataset for the analysis
# Merge the two datasets 'customer_transaction' & 'cab'

cab_cust_trans = pd.merge(cab, customer_transaction)
cab_cust_trans

#Now, merge the 4th dataset 'City' to get all in one DATASET on which all the analysis is done

df = pd.merge(cab_cust_trans, city)
df

#The final dataset df contains 359392 rows/individuals and 14 columns/variables
```

```python
# Correlation plot
# relation between the different variables of the dataset

plt.figure(figsize=(10,20))
c = df.corr()
sns.heatmap(c, 
cmap='RdBu_r', 
annot=True,
vmin=-1, vmax=1)
c
```

```python
# analysis based on the charged price
sns.barplot(data=df, x="Company", y="Price Charged")
## The Yellow Cab company charges more than the Pink Cab company

#analysis based on cost of the trip
sns.barplot(data=df, x="Company", y="Cost of Trip")
## Since the Price Charged and Cost of Trip are proportional and highly correlated 
## Then the Yellow Cab company's trips cost more than those of the Pink Cab company
```

```python
# Bivariate analysis
# previous analysis confirmation
# based on both price charged and cost of trip
sns.relplot(
    x='Price Charged', 
    y='Cost of Trip', 
    data=df, 
    kind='scatter', 
    col='Company');

## Yellow cab company both charges and costs more than the Pink cab company.
```

```python
# Multivariate analysis
sns.catplot(
    x='Company', 
    y='Cost of Trip', 
    data=df, 
    palette='bright',
    kind='point', 
    hue='Payment_Mode');

## The distribution of the payment method between card cash is almost the same for  the Yellow cab firm and the Pink cab firm.
```

```python
sns.barplot(data=df, x="Company", y="Age")

# Equal distribution of age among the two cab companies Pink Cab & Yellow Cab.
```
